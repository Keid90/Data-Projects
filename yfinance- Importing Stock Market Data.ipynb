```python
import yfinance as yf
import matplotlib.pyplot as plt

nvidia = yf.Ticker("NVDA")
data = nvidia.history(period="1y")  # Get past year's data

# Plot the closing prices
plt.figure(figsize=(8, 4))
plt.plot(data.index, data["Close"], linestyle="-", label="NVIDIA Share Price")

# Formatting the plot
plt.xlabel("Date")
plt.ylabel("Share Price (USD)")
plt.title("NVIDIA Share Price Over the Last Year")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
